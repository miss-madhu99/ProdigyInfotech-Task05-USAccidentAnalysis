```python
import pandas as pd
import numpy as np
import plotly.express as px
n = 1000
df = pd.DataFrame({
    'Latitude': np.random.uniform(8.0, 15.0, n),
    'Longitude': np.random.uniform(38.0, 45.0, n),
    'Accident_severity': np.random.choice(['Slight', 'Serious', 'Fatal'], n)
})
fig = px.scatter_geo(
    df,
    lat='Latitude',
    lon='Longitude',
    color='Accident_severity',
    title='Accident Map (Plotly Version)',
    opacity=0.6,
    template='plotly_dark'
)
fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({
    'Severity': np.random.randint(1, 5, size=1000),
    'Visibility(mi)': np.random.uniform(0.1, 10, size=1000),
    'Humidity(%)': np.random.uniform(20, 100, size=1000),
    'Temperature(F)': np.random.uniform(30, 100, size=1000),
    'Wind_Speed(mph)': np.random.uniform(0, 20, size=1000)
})

# Calculate correlation
corr = df[['Severity', 'Visibility(mi)', 'Humidity(%)', 'Temperature(F)', 'Wind_Speed(mph)']].corr()

# Plotting heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Severity and Weather Factors")
plt.tight_layout()
plt.show()
```
